```python
import pandas as pd
```

```python
df = pd.read_csv("../data/retail_sales_dataset.csv")
# to Load raw dataset
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# to Convert Date column
```

```python
df.drop_duplicates(inplace=True)
#  to Remove duplicate rows
```

```python
df['Estimated_Profit'] = df['Total Amount'] * 0.25
df['Profit_Margin'] = df['Estimated_Profit'] / df['Total Amount']
# feature
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 25, 35, 50, 100],
    labels=['18-25', '26-35', '36-50', '50+'] )
# TO Create Age Groups
```

```python
print(df.head())
print(df.isnull().sum())
# final check
```

```python
df.to_csv("../data/clean_retail_sales.csv", index=False)
# Save cleaned dataset
```
